```python
np.random.seed(613)
x1=np.arange(50)
y1=np.random.randint(0,75,50)
x2=np.array(["Luff","Zoro","Nami","Usopp","Sanji"])
y2=np.array([110,180,240,99,220])
```

```python
# Plot the 2 graphs
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize = (7,3))

ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
# Define different colors for each bar
bar_colors = ['orange', 'green', 'blue', 'purple', 'red']

# Experiment with formatting
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(10, 4),
                               )

# Scatter plot formatting
ax1.scatter(x1, y1, color="orange")
ax1.set_title('Random Test Scores', fontsize=20, fontweight='bold', color='navy')
ax1.set_xlabel('Test Numbers', fontsize=14, fontfamily='serif', color='darkred')
ax1.set_ylabel('Scores', fontsize=14, fontfamily='serif', color='darkred')
ax1.grid(True, linestyle='--', alpha=0.6)

# Bar plot formatting with different colors for each bar
ax2.bar(x2, y2, color=bar_colors)
ax2.set_title('Character Scores', fontsize=20, fontweight='bold', color='navy')
ax2.set_xlabel('Names', fontsize=14, fontfamily='serif', color='darkred')
ax2.set_ylabel('Scores', fontsize=14, fontfamily='serif', color='darkred')
ax2.grid(True, linestyle='--', alpha=0.6)

# Common formatting for both plots
for ax in [ax1, ax2]:
    ax.tick_params(axis='both', which='major', labelsize=12)
    ax.set_facecolor('whitesmoke')

# Add annotations
for i, value in enumerate(y1):
    ax1.annotate(value,
                 (x1[i], y1[i]),
                 textcoords="offset points",
                 xytext=(0, 5),
                 ha='center',
                 fontsize=10,
                 color='darkgreen'
                 )

for i, value in enumerate(y2):
    ax2.text(x2[i], value + 10, str(value), ha='center', va='center', fontsize=12, color='darkgreen')

plt.tight_layout()
plt.show()
```

```python
##first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

##define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))

##now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,
      color = "indigo")
ax.plot(x, y2,
       color = "red")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x, #our x values
            y2, #our y values
           yerr = y2_sd, 
           fmt = "none")

```

```python
from PIL import Image ##to open images
import requests ##to get images from URLs
from io import BytesIO ##to store images
```

```python
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax_image = fig.add_axes([0.1, ##x coordinate (ON FIGURE, NOT AXES)
                         0.11, ##y coordinate (ON FIGURE, NOT AXES)
                         0.15, ##image width
                         0.35] )##image height)

ax_image.imshow(image)
ax_image.axis('off')
plt.show()
```
